```python
import numpy as np
from PIL import Image
```

```python
img = Image.open("test.png")
try:
    data = np.asarray( img, dtype='uint8' )
except SystemError:
    data = np.asarray( img.getdata(), dtype='uint8' )
```

```python
chunked = []
for i in range(10):
    chunked.append(data[i * 3:i * 3 + 3, i * 3:i * 3 + 3, :3])
```

```python
print(chunked)
```

```python
tests = []
for chunk in chunked:
    test = ""
    for column in chunk:
        for pixel in column:
            test += format(int(pixel[0]), '#010b') + format(int(pixel[1]), '#010b') + format(int(pixel[2]), '#010b')
    test = test.replace("0b", "")
    tests.append(test)
```
